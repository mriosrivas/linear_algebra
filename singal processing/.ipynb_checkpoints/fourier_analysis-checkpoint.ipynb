```python
import numpy as np
```

```python
from scipy.io import wavfile
```

```python
[rate, song] = wavfile.read('Perfect Strangers.wav')
```

```python
left_channel = song[:,0]
right_channel = song[:,1]
```

```python
import matplotlib.pyplot as plt
```

```python
samples = 75
left = left_channel[0:samples]
right = right_channel[0:samples]

plt.rcParams["figure.figsize"] = (10,5)
plt.subplot(2,2,1)
plt.stem(left, use_line_collection=True);
plt.title("Left Channel")
plt.subplot(2,2,2)
plt.stem(right, use_line_collection=True);
plt.title("Right Channel");
plt.subplot(2,1,2)
plt.stem(left+right, use_line_collection=True);
plt.title("Mixed Channel");
```

```python
frequency = np.linspace(0,rate/2,samples)
left_fft = np.fft.fft(left)
right_fft = np.fft.fft(right)
mixed = left_fft + right_fft

plt.rcParams["figure.figsize"] = (10,5)
plt.subplot(2,2,1)
plt.stem(frequency, np.abs(left_fft), use_line_collection=True);
plt.title("Left Channel")
plt.subplot(2,2,2)
plt.stem(frequency, np.abs(right_fft), use_line_collection=True);
plt.title("Right Channel");
plt.subplot(2,1,2)
plt.stem(np.abs(mixed), use_line_collection=True);
plt.title("Mixed Channel");
```

```python
mixed_ifft = np.real(np.fft.ifft(mixed))
```

```python
plt.subplot(2,1,1)
plt.stem(left+right, use_line_collection=True);
plt.title("Mixed Channel");
plt.subplot(2,1,2)
plt.stem(mixed_ifft, use_line_collection=True);
plt.title("Mixed Channel IFFT");
```

```python
left_fft = np.fft.fft(left_channel)
right_fft = np.fft.fft(right_channel)
mixed = left_fft + right_fft
mixed_ifft = np.real(np.fft.ifft(mixed))
```

```python
wavfile.write('Sum Perfect Strangers.wav', rate, (left_channel+right_channel)/100)
wavfile.write('Sum IFFT Perfect Strangers.wav', rate, mixed_ifft/100)
```
